```python
import os
import cv2
import numpy as np
import random
import matplotlib.pyplot as plt
from collections import Counter
from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay, f1_score, roc_curve
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout, Input
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from tensorflow.keras.optimizers import Adam
```

```python
IMG_SIZE = 224
BATCH_SIZE = 32
DATADIR = r"C:\Users\ayush\OneDrive\Desktop\project3\chest_xray"
```

```python
def load_data(directory):
    data = []
    for label in ["NORMAL", "PNEUMONIA"]:
        path = os.path.join(directory, label)
        class_num = 0 if label == "NORMAL" else 1
        for img_name in os.listdir(path):
            try:
                img_path = os.path.join(path, img_name)
                img = cv2.imread(img_path)
                img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
                data.append([img, class_num])
            except:
                continue
    return data
```

```python
data = load_data(os.path.join(DATADIR, r"C:\Users\ayush\OneDrive\Desktop\project3\chest_xray\train")) + \
       load_data(os.path.join(DATADIR, r"C:\Users\ayush\OneDrive\Desktop\project3\chest_xray\val")) + \
       load_data(os.path.join(DATADIR, r"C:\Users\ayush\OneDrive\Desktop\project3\chest_xray\test"))
```

```python
X, y = zip(*data)
X = np.array(X)
y = np.array(y)

# Stratified train-val-test split (80-10-10)
X_temp, X_test, y_temp, y_test = train_test_split(X, y, test_size=0.1, stratify=y, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.1111, stratify=y_temp, random_state=42)  # 0.1111 * 0.9 ≈ 0.1
```

```python
X_train = preprocess_input(X_train)
X_val = preprocess_input(X_val)
X_test = preprocess_input(X_test)
```

```python
train_datagen = ImageDataGenerator(
    rotation_range=15,
    zoom_range=0.2,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    horizontal_flip=True
)
val_datagen = ImageDataGenerator()

train_generator = train_datagen.flow(X_train, y_train, batch_size=BATCH_SIZE)
val_generator = val_datagen.flow(X_val, y_val, batch_size=BATCH_SIZE)
```

```python
class_weights = compute_class_weight(class_weight='balanced', classes=np.unique(y_train), y=y_train)
class_weight_dict = dict(enumerate(class_weights))
```

```python
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(IMG_SIZE, IMG_SIZE, 3))
base_model.trainable = False  # freeze base

x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dropout(0.5)(x)
output = Dense(1, activation='sigmoid')(x)

model = Model(inputs=base_model.input, outputs=output)

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', patience=2, factor=0.2, verbose=1)
```

```python
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=10,
    class_weight=class_weight_dict,
    callbacks=[early_stop, reduce_lr]
)
```

```python

base_model.trainable = True
model.compile(optimizer=Adam(1e-5), loss='binary_crossentropy', metrics=['accuracy'])

#  STEP 3: Fine-tune the full model
history_finetune = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=5,  # you can increase this if needed
    class_weight=class_weight_dict,
    callbacks=[early_stop, reduce_lr]
)
```

```python
model.save("pneumonia_mobilenetv2_finetuned.keras")
print("Fine-tuned model saved as 'pneumonia_mobilenetv2_finetuned.keras'")
```

```python
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {accuracy * 100:.2f}%")
```

```python
plt.plot(history.history['accuracy'], label='Initial Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Initial Val Accuracy')
plt.plot(history_finetune.history['accuracy'], label='Fine-Tune Train Accuracy')
plt.plot(history_finetune.history['val_accuracy'], label='Fine-Tune Val Accuracy')
plt.title("Model Accuracy Over Epochs")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.show()
```

```python
y_probs = model.predict(X_test)
```

```python
# F1-based threshold tuning
best_thresh = 0.5
best_f1 = 0.0
for t in np.arange(0.3, 0.8, 0.01):
    y_pred = (y_probs > t).astype("int32")
    f1 = f1_score(y_test, y_pred)
    if f1 > best_f1:
        best_f1 = f1
        best_thresh = t
print(f"\nBest Threshold by F1-score: {best_thresh:.2f} (F1 = {best_f1:.4f})")
```

```python
fpr, tpr, roc_thresholds = roc_curve(y_test, y_probs)
j_scores = tpr - fpr
j_best_index = np.argmax(j_scores)
optimal_roc_threshold = roc_thresholds[j_best_index]
print(f"Best Threshold by ROC (Youden’s J): {optimal_roc_threshold:.2f}")
```

```python
y_pred = (y_probs > best_thresh).astype("int32")

print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=["NORMAL", "PNEUMONIA"]))

cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["NORMAL", "PNEUMONIA"])
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix (Fine-Tuned + Best Threshold)")
plt.show()
```

```python
# -------------------- Visualize Multiple Random Test Predictions --------------------
label_map = {0: "NORMAL", 1: "PNEUMONIA"}

num_samples = 20  # You can change this number

for i in range(num_samples):
    index = random.randint(0, len(X_test) - 1)
    image = X_test[index]
    actual_label = y_test[index]

    # Predict
    input_img = np.expand_dims(image, axis=0)
    prediction = model.predict(input_img)[0][0]
    predicted_label = 1 if prediction >= best_thresh else 0
    confidence = prediction if predicted_label == 1 else 1 - prediction

    # Rescale image from [-1, 1] to [0, 255] for display
    img_to_display = (image + 1) / 2
    img_rgb = cv2.cvtColor((img_to_display * 255).astype(np.uint8), cv2.COLOR_BGR2RGB)

    # Show
    plt.figure(figsize=(4, 4))
    plt.imshow(img_rgb)
    plt.title(f"Prediction: {label_map[predicted_label]} ({confidence*100:.2f}%)\nActual: {label_map[actual_label]}")
    plt.axis('off')
    plt.tight_layout()
    plt.show()


```
